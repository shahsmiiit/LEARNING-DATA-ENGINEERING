## 1. Sorting DataFrames

````python
from pyspark.sql.functions import col, desc

# Create sample data
data = [
    ("Alice", 28, 70000),
    ("Bob", 32, 85000),
    ("Charlie", 25, 60000),
    ("Diana", 35, 95000),
    ("Eve", 29, 75000)
]
df = spark.createDataFrame(data, ["name", "age", "salary"])

print("Original data:")
df.show()
````

**Output:**
```
+-------+---+------+
|   name|age|salary|
+-------+---+------+
|  Alice| 28| 70000|
|    Bob| 32| 85000|
|Charlie| 25| 60000|
|  Diana| 35| 95000|
|    Eve| 29| 75000|
+-------+---+------+
```

````python
# Sort by single column (ascending)
print("Sort by age (ascending):")
df.sort("age").show()
````

**Output:**
```
+-------+---+------+
|   name|age|salary|
+-------+---+------+
|Charlie| 25| 60000|
|  Alice| 28| 70000|
|    Eve| 29| 75000|
|    Bob| 32| 85000|
|  Diana| 35| 95000|
+-------+---+------+
```

````python
# Sort descending
print("Sort by salary (descending):")
df.sort(col("salary").desc()).show()
````

**Output:**
```
+-----+---+------+
| name|age|salary|
+-----+---+------+
|Diana| 35| 95000|
|  Bob| 32| 85000|
|  Eve| 29| 75000|
|Alice| 28| 70000|
+-----+---+------+
```

````python
# Sort by multiple columns
print("Sort by age ascending, then salary descending:")
df.sort(col("age").asc(), col("salary").desc()).show()
````

**Output:**
```
+-------+---+------+
|   name|age|salary|
+-------+---+------+
|Charlie| 25| 60000|
|  Alice| 28| 70000|
|    Eve| 29| 75000|
|    Bob| 32| 85000|
|  Diana| 35| 95000|
+-------+---+------+
```

````python
# Limit results
print("Top 3 by salary:")
df.sort(col("salary").desc()).limit(3).show()
````

**Output:**
```
+-----+---+------+
| name|age|salary|
+-----+---+------+
|Diana| 35| 95000|
|  Bob| 32| 85000|
|  Eve| 29| 75000|
+-----+---+------+
```

## 2. Joining DataFrames

Combine data from multiple DataFrames.

````python
# Create two DataFrames
employees = [
    (1, "Alice", "Engineering"),
    (2, "Bob", "Sales"),
    (3, "Charlie", "Engineering")
]
df_emp = spark.createDataFrame(employees, ["emp_id", "name", "department"])

salaries = [
    (1, 70000),
    (2, 85000),
    (4, 90000)  # Employee 4 doesn't exist in employees
]
df_sal = spark.createDataFrame(salaries, ["emp_id", "salary"])

print("Employees:")
df_emp.show()
print("\nSalaries:")
df_sal.show()
````

**Output:**
```
Employees:
+------+-------+----------+
|emp_id|   name|department|
+------+-------+----------+
|     1|  Alice|Engineering|
|     2|    Bob|     Sales|
|     3|Charlie|Engineering|
+------+-------+----------+

Salaries:
+------+------+
|emp_id|salary|
+------+------+
|     1| 70000|
|     2| 85000|
|     4| 90000|
+------+------+
```

````python
# Inner join (only matching rows)
print("INNER JOIN:")
df_inner = df_emp.join(df_sal, "emp_id", "inner")
df_inner.show()
````

**Output:**
```
+------+-----+----------+------+
|emp_id| name|department|salary|
+------+-----+----------+------+
|     1|Alice|Engineering| 70000|
|     2|  Bob|     Sales| 85000|
+------+-----+----------+------+
```

````python
# Left outer join (keep all from left)
print("LEFT OUTER JOIN:")
df_left = df_emp.join(df_sal, "emp_id", "left")
df_left.show()
````

**Output:**
```
+------+-------+----------+------+
|emp_id|   name|department|salary|
+------+-------+----------+------+
|     1|  Alice|Engineering| 70000|
|     2|    Bob|     Sales| 85000|
|     3|Charlie|Engineering|  null|
+------+-------+----------+------+
```

````python
# Right outer join (keep all from right)
print("RIGHT OUTER JOIN:")
df_right = df_emp.join(df_sal, "emp_id", "right")
df_right.show()
````

**Output:**
```
+------+-----+----------+------+
|emp_id| name|department|salary|
+------+-----+----------+------+
|     1|Alice|Engineering| 70000|
|     2|  Bob|     Sales| 85000|
|     4| null|      null| 90000|
+------+-----+----------+------+
```

````python
# Full outer join (keep all from both)
print("FULL OUTER JOIN:")
df_full = df_emp.join(df_sal, "emp_id", "outer")
df_full.show()
````

**Output:**
```
+------+-------+----------+------+
|emp_id|   name|department|salary|
+------+-------+----------+------+
|     1|  Alice|Engineering| 70000|
|     2|    Bob|     Sales| 85000|
|     3|Charlie|Engineering|  null|
|     4|  null|      null| 90000|
+------+-------+----------+------+
```

## 3. Union Operations

Combine data vertically (stack rows).

### Union vs UnionByName vs Except vs Intersect

# Create two DataFrames with same schema
data1 = [("Alice", 70000), ("Bob", 85000)]
data2 = [("Charlie", 60000), ("Diana", 95000)]

df1 = spark.createDataFrame(data1, ["name", "salary"])
df2 = spark.createDataFrame(data2, ["name", "salary"])

print("DataFrame 1:")
df1.show()
print("\nDataFrame 2:")
df2.show()

# Union (combine all rows)
print("\nUNION (all rows):")
df_union = df1.union(df2)
df_union.show()

**Output:**
```
DataFrame 1:
+-----+------+
| name|salary|
+-----+------+
|Alice| 70000|
|  Bob| 85000|
+-----+------+

DataFrame 2:
+-------+------+
|   name|salary|
+-------+------+
|Charlie| 60000|
|  Diana| 95000|
+-------+------+

UNION (all rows):
+-------+------+
|   name|salary|
+-------+------+
|  Alice| 70000|
|    Bob| 85000|
|Charlie| 60000|
|  Diana| 95000|
+-------+------+
```

# Union removes duplicates, unionByName keeps them
data3 = [("Alice", 70000), ("Bob", 85000)]
data4 = [("Alice", 70000), ("Charlie", 60000)]

df3 = spark.createDataFrame(data3, ["name", "salary"])
df4 = spark.createDataFrame(data4, ["name", "salary"])

print("Union (removes duplicates):")
df3.union(df4).show()

print("\nUnionByName (keeps duplicates):")
df3.unionByName(df4).show()

**Output:**
```
Union (removes duplicates):
+-------+------+
|   name|salary|
+-------+------+
|  Alice| 70000|
|    Bob| 85000|
|Charlie| 60000|
+-------+------+

UnionByName (keeps duplicates):
+-------+------+
|   name|salary|
+-------+------+
|  Alice| 70000|
|    Bob| 85000|
|  Alice| 70000|
|Charlie| 60000|
+-------+------+
```

# Set operations
print("INTERSECT (common rows):")
df3.intersect(df4).show()

print("\nEXCEPT (in df3 but not df4):")
df3.exceptAll(df4).show()

**Output:**
```
INTERSECT (common rows):
+-----+------+
| name|salary|
+-----+------+
|Alice| 70000|
+-----+------+

EXCEPT (in df3 but not df4):
+---+------+
|name|salary|
+---+------+
|Bob| 85000|
+---+------+
```
