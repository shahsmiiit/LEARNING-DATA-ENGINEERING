## 1. Displaying DataFrames

Before processing data, you need to explore and understand it. Let's start with different display methods.

````python
# Create a sample DataFrame
data = [
    ("Alice", 28, "Engineering", 70000),
    ("Bob", 32, "Sales", 85000),
    ("Charlie", 25, "Marketing", 60000),
    ("Diana", 35, "Engineering", 95000),
    ("Eve", 29, "Sales", 75000)
]
columns = ["name", "age", "department", "salary"]
df = spark.createDataFrame(data, schema=columns)

# Method 1: show() - Default display
print("Method 1: show()")
df.show()
````

**Output:**
```
+-------+---+----------+------+
|   name|age|department|salary|
+-------+---+----------+------+
|  Alice| 28|Engineering| 70000|
|    Bob| 32|      Sales| 85000|
|Charlie| 25|  Marketing| 60000|
|  Diana| 35|Engineering| 95000|
|    Eve| 29|      Sales| 75000|
+-------+---+----------+------+
```

````python
# Method 2: show() with custom number of rows
print("Method 2: show(3) - Show only 3 rows")
df.show(3)
````

**Output:**
```
+-------+---+----------+------+
|   name|age|department|salary|
+-------+---+----------+------+
|  Alice| 28|Engineering| 70000|
|    Bob| 32|      Sales| 85000|
|Charlie| 25|  Marketing| 60000|
+-------+---+----------+------+
only showing top 3 rows
```

````python
# Method 3: show() with truncation disabled
print("Method 3: show(truncate=False) - Don't truncate long columns")
df.show(truncate=False)
````

**Output:**
```
+-------+---+----------+------+
|name   |age|department|salary|
+-------+---+----------+------+
|Alice  | 28|Engineering| 70000|
|Bob    | 32|Sales      | 85000|
|Charlie| 25|Marketing  | 60000|
|Diana  | 35|Engineering| 95000|
|Eve    | 29|Sales      | 75000|
+-------+---+----------+------+
```

````python
# Method 4: collect() - Convert to Python list (CAREFUL: loads all data to driver)
print("Method 4: collect() - Get all data as Python objects")
rows = df.collect()
print(f"Number of rows: {len(rows)}")
print(f"First row: {rows[0]}")
print(f"First row name: {rows[0].name}")
````

**Output:**
```
Number of rows: 5
First row: Row(name='Alice', age=28, department='Engineering', salary=70000)
First row name: Alice
```

⚠️ **Warning:** `collect()` loads ALL data into memory on the driver. Don't use on huge DataFrames!

````python
# Method 5: first() - Get only first row
print("Method 5: first() - Get first row only")
first_row = df.first()
print(f"First row: {first_row}")
print(f"Age of first person: {first_row['age']}")
````

**Output:**
```
First row: Row(name='Alice', age=28, department='Engineering', salary=70000)
Age of first person: 28
```

````python
# Method 6: take() - Get first N rows
print("Method 6: take(2) - Get first 2 rows as list")
rows = df.take(2)
for row in rows:
    print(row)
````

**Output:**
```
Row(name='Alice', age=28, department='Engineering', salary=70000)
Row(name='Bob', age=32, department='Sales', salary=85000)
```

## 4. Renaming Columns

Change column names for clarity or data standardization.

````python
# Method 1: withColumnRenamed() - Rename single column
print("Method 1: Rename 'salary' to 'annual_salary'")
df_renamed = df.withColumnRenamed('salary', 'annual_salary')
df_renamed.show(3)
````

**Output:**
```
+-------+---+----------+-------------+
|   name|age|department|annual_salary|
+-------+---+----------+-------------+
|  Alice| 28|Engineering|        70000|
|    Bob| 32|      Sales|        85000|
|Charlie| 25|  Marketing|        60000|
+-------+---+----------+-------------+
```

````python
# Method 2: Multiple renames (chain them)
print("Method 2: Rename multiple columns")
df_renamed = df \
    .withColumnRenamed('name', 'employee_name') \
    .withColumnRenamed('age', 'employee_age') \
    .withColumnRenamed('salary', 'annual_salary')
df_renamed.show(3)
````

**Output:**
```
+-------------+------------+----------+-------------+
|employee_name|employee_age|department|annual_salary|
+-------------+------------+----------+-------------+
|        Alice|          28|Engineering|        70000|
|          Bob|          32|      Sales|        85000|
|      Charlie|          25|  Marketing|        60000|
+-------------+------------+----------+-------------+
```

````python
# Method 3: Rename all columns at once
print("Method 3: Rename all columns")
new_names = ['emp_name', 'emp_age', 'dept', 'salary']
df_renamed = df.toDF(*new_names)  # Unpack list as arguments
df_renamed.show(3)
````

**Output:**
```
+--------+-------+----------+------+
|emp_name|emp_age|      dept|salary|
+--------+-------+----------+------+
|   Alice|     28|Engineering| 70000|
|     Bob|     32|     Sales| 85000|
|  Charlie|     25|  Marketing| 60000|
+--------+-------+----------+------+
```

## 5. Basic Statistics and Aggregations

Get quick insights about your data.

````python
# Method 1: describe() - Show statistics
print("Method 1: describe() - Basic statistics")
df.describe().show()
````

**Output:**
```
+-------+-------+----------+------+
|summary|    age|salary    |salary|
+-------+-------+----------+------+
|  count|      5|         5|     5|
|   mean|     29.8|        77000.0|
| stddev|3.9621|15684.387|
|    min|     25|        60000|
|    max|     35|        95000|
+-------+-------+----------+------+
```

````python
# Method 2: describe() - For specific columns
print("Method 2: describe('age', 'salary')")
df.describe('age', 'salary').show()
````

**Output:**
```
+-------+---+------+
|summary|age|salary|
+-------+---+------+
|  count|  5|     5|
|   mean|29.8|77000|
| stddev|3.96|15684.387|
|    min| 25| 60000|
|    max| 35| 95000|
+-------+---+------+
```

````python
# Method 3: Use functions for specific aggregations
from pyspark.sql.functions import min, max, avg, stddev, count

print("Method 3: Custom aggregations")
df.agg(
    count('age').alias('count'),
    min('age').alias('min_age'),
    max('age').alias('max_age'),
    avg('age').alias('avg_age'),
    avg('salary').alias('avg_salary')
).show()
````

**Output:**
```
+-----+-------+-------+-------+----------+
|count|min_age|max_age|avg_age|avg_salary|
+-----+-------+-------+-------+----------+
|    5|     25|     35|   29.8|      77000|
+-----+-------+-------+-------+----------+
```

## 6. Distinct Values and Unique Count

````python
# Create sample data with duplicates
data_with_dups = [
    ("Alice", "Engineering"),
    ("Bob", "Sales"),
    ("Charlie", "Engineering"),
    ("Diana", "Engineering"),
    ("Eve", "Sales")
]
df_depts = spark.createDataFrame(data_with_dups, ["name", "department"])

# Method 1: Get distinct values
print("Method 1: distinct() - Get unique rows")
df_depts.select('department').distinct().show()
````

**Output:**
```
+----------+
|department|
+----------+
|Engineering|
|     Sales|
+----------+
```

````python
# Method 2: Count distinct values
from pyspark.sql.functions import countDistinct

print("Method 2: countDistinct()")
df_depts.agg(countDistinct('department')).show()
````

**Output:**
```
+-----+
|count(department)|
+-----+
|    2|
+-----+
```

## 7. Null Values Check

Understanding missing data is critical.

````python
# Create data with nulls
data_with_nulls = [
    ("Alice", 28, "Engineering"),
    ("Bob", None, "Sales"),
    ("Charlie", 25, None),
    ("Diana", 35, "Engineering"),
    (None, 29, "Sales")
]
df_nulls = spark.createDataFrame(data_with_nulls, ["name", "age", "department"])

print("DataFrame with null values:")
df_nulls.show()
````

**Output:**
```
+-------+----+----------+
|   name| age|department|
+-------+----+----------+
|  Alice|  28|Engineering|
|    Bob|null|     Sales|
|Charlie|  25|      null|
|  Diana|  35|Engineering|
|   null|  29|     Sales|
+-------+----+----------+
```

````python
# Check for nulls
from pyspark.sql.functions import isNull, isNotNull, when, sum as spark_sum

print("Count of null values in each column:")
df_nulls.agg(
    spark_sum(when(isNull('name'), 1).otherwise(0)).alias('null_names'),
    spark_sum(when(isNull('age'), 1).otherwise(0)).alias('null_ages'),
    spark_sum(when(isNull('department'), 1).otherwise(0)).alias('null_departments')
).show()
````

**Output:**
```
+----------+----------+------------------+
|null_names|null_ages|null_departments|
+----------+----------+------------------+
|         1|         1|                1|
+----------+----------+------------------+
```

## 8. Comprehensive Data Profile

Create a function to get complete data overview:

````python
def profile_dataframe(df):
    """Print comprehensive DataFrame profile"""
    print("="*60)
    print("DATAFRAME PROFILE")
    print("="*60)
    
    # Basic info
    print(f"\n✓ Number of rows: {df.count()}")
    print(f"✓ Number of columns: {len(df.columns)}")
    print(f"✓ Column names: {df.columns}")
    
    # Schema
    print("\n✓ Schema:")
    df.printSchema()
    
    # Sample data
    print("\n✓ Sample data:")
    df.show(3, truncate=False)
    
    # Statistics
    print("\n✓ Statistics:")
    df.describe().show()
    
    print("="*60)

# Test it
profile_dataframe(df)
````
